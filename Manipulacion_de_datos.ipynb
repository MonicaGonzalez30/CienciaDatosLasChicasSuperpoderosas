```python
def quick_sort_pirata(arreglo):
    izq = []
    der = []
    center = []
    
    if len(arreglo) == 0 or len(arreglo)==1:
        return arreglo
    
    else:
        
        piv = arreglo[0]
        
        for e in arreglo:
            if e<piv:
                izq.append(e)
            elif e==piv:
                center.append(e)
            elif e>piv:
                der.append(e)
        
        return quick_sort_pirata(izq)+center+quick_sort_pirata(der)
    

def merge_sort_pirata(arreglo):
    # Casos base
    if len(arreglo) == 0 or len(arreglo) == 1:
        return arreglo
    
    else:
        mitad = len(arreglo) // 2
        arr_izq = arreglo[:mitad]
        arr_der = arreglo[mitad:]
        
        arr_izq = merge_sort_pirata(arr_izq)
        arr_der = merge_sort_pirata(arr_der)
        
        a = 0
        b = 0
        c = 0
        
        while a < len(arr_izq) and b < len(arr_der):
            if arr_izq[a] < arr_der[b]:
                arreglo[c] = arr_izq[a]
                a += 1
            else:
                arreglo[c] = arr_der[b]
                b += 1
            c += 1
        
        while a < len(arr_izq):
            arreglo[c] = arr_izq[a]
            a += 1
            c += 1
        
        while b < len(arr_der):
            arreglo[c] = arr_der[b]
            b += 1
            c += 1
            
        return arreglo
```
